## how **__init__.py** works in top package?
- it helps to think of it as the "Activation Switch" for your package.
- It doesn't run during installation, but rather at the exact moment your code tries to "touch" the package.

1. **Does it start first when I run a script?**
- No, not automatically. If you run a standalone script like python scripts/train.py, the __init__.py inside src/deep_learning_package/ will only execute if train.py contains a line that imports something from that package.

2. **Does it start with importing any module from the package?**
- Yes! This is exactly when it runs. The very first time any module in your code (or a cell in your Notebook) executes an import statement targeting that package, Python executes the code inside __init__.py from top to bottom.

- Example: If you run from deep_learning_package.architecture import DeepModel, Python first runs deep_learning_package/__init__.py to initialize the namespace, then it proceeds to load architecture.py.

- One-time Execution: It only runs once per session. If you import another module later, Python remembers it has already initialized the package and won't run __init__.py again.

3. **Does it run during pip install -e .?**
- No. The pip install -e . command (driven by your pyproject.toml) only creates the links between your Python environment and your source code. It checks your metadata and dependencies, but it does not execute your package's logic.


## **Comparison Summary**
|Event |Does __init__.py run?| Why?|
|:---| :---| :---| 
|**pip install -e .**| No | Installation only maps the file paths; it doesn't execute code|
|import deep_learning_package| Yes| Python must initialize the package to see what is inside.|
|Running a separate script | Only if it imports the package| Scripts are independent until they call for the package code.|

# Mapping Biological Neurons to AI Engineering

This document provides a one-line engineering explanation for each component of a biological neuron and its equivalent in Artificial Neural Networks (ANN).

### **The AI Component Map**

| Biological Component | AI Engineering Equivalent | Functional Explanation (One Line) |
| :--- | :--- | :--- |
| **Dendrites** | **Input Features ($x$)** | Receives raw data or signals from previous layers. |
| **Synapses** | **Weights ($w$)** | Scales the input signal to represent its relative importance. |
| **Soma** | **Summation Node ($\sum$)** | Aggregates all weighted inputs and adds the bias term. |
| **Axon Hillock** | **Activation Threshold** | Determines if the aggregated signal is sufficient to "fire." |
| **Axon** | **Output Propagation** | Transmits the activated signal to the next set of neurons. |
| **Myelin Sheath** | **Optimization/Norms** | Enhances signal speed and reduces vanishing gradients/noise. |
| **Axon Terminals** | **Layer Output** | Forwards the processed signal to multiple destination nodes. |

### **The Mathematical Flow**
In AI engineering, the entire process described above is compressed into a single mathematical operation per neuron:
$$Output = Activation(\sum(Weights \times Inputs) + Bias)$$

### **Why This Matters**
Understanding this mapping helps AI engineers design better architectures. For example, **Activation Functions** (the Axon's logic) prevent the network from being a simple linear calculator, allowing it to learn complex patterns like faces and human speech.

### Regression/Linear Model 
The function implemented by a neuron with no activation is the same as in Course 1, linear regression:
$$ f_{\mathbf{w},b}(x^{(i)}) = \mathbf{w}\cdot x^{(i)} + b \tag{1}$$



Defined a layer with one neuron (in keras: units = 1 --> one neuron) and compare it to the familiar linear regression function.

````python
linear_layer = Dense(units = 1,input_dim = 1, activation='linear',name = 'Layer1')
````

- Type: The object **linear_layer** is of type ``tf.keras.layers.Dense``. In programming terms, it is an **"instance"** of the ``Dense`` class.

- Values: Internally, this layer holds **two variables**: ``Kernel ($w$): The weight (slope)`` and ``Bias ($b$): The intercept``.

- Note: When you first create the layer, these values are initialized with ``random small numbers (or zeros for bias)``. 

- They only become "correct" after you **train the model with data**.

1. What is **linear_layer.get_weights()**? 
    - This is the most important command for "looking inside" your model. 
    - It returns a Python list containing two NumPy arrays. 
    - get_weights()[0]: This is the Weight ($w$). For your specific code, it will be a 2D array of shape (1, 1).
    - get_weights()[1]: This is the Bias ($b$). It will be a 1D array of shape (1,).

````python
# Extract the values
weights_bias = linear_layer.get_weights()

print(f"The weight_bias (w) is: {weights_bias}")
````

There are no weights as the weights are not yet instantiated. Let's try the model on one example in `X_train`. This will trigger the instantiation of the weights. Note, the input to the layer must be 2-D, so we'll reshape it.

````python
a1 = linear_layer(X_train[0].reshape(1,1))

print(a1)
````

Why did you have to use .reshape(1,1)?
TensorFlow layers are designed to process batches of data, not just one single number.

If you just passed X_train[0] (a single scalar), the model would get confused.

By using .reshape(1,1), you are telling the model: "Here is a batch of 1 item, and that item has 1 feature."

````python
w, b= linear_layer.get_weights()
print(f"w = {w}, b={b}")
````

A linear regression model (1) with a single input feature will have a single weight and bias. This matches the dimensions of our `linear_layer` above.   

The weights are initialized to random values so let's set them to some known values.

````python
set_w = np.array([[200]])
set_b = np.array([100])

# set_weights takes a list of numpy arrays
linear_layer.set_weights([set_w, set_b])
print(linear_layer.get_weights())
````

````python
a1 = linear_layer(X_train[0].reshape(1,1))
print(a1)
alin = np.dot(set_w,X_train[0].reshape(1,1)) + set_b
print(alin)
````

# Linear Model Training & Prediction Summary

This table summarizes the end-to-end workflow of implementing a Linear Regression model using a `Dense` layer in TensorFlow.

| Step | Function | Engineering Purpose | Example Code Snippet |
| :--- | :--- | :--- | :--- |
| **1. Define** | `Sequential()` & `Dense()` | Sets the architecture (inputs, neurons, and activation). | `model = Sequential([Dense(units=1, input_dim=1)])` |
| **2. Compile** | `model.compile()` | Defines the "Rules" (Optimizer for updates, Loss to measure error). | `model.compile(optimizer='adam', loss='mse')` |
| **3. Train** | `model.fit()` | The "Learning" phase where Weights ($w$) and Bias ($b$) are optimized. | `model.fit(X_train, y_train, epochs=500)` |
| **4. Predict** | `model.predict()` | Uses the learned $w$ and $b$ to calculate results for new data. | `y_pred = model.predict([[5.0]])` |
| **5. Inspect** | `get_weights()` | Extracts the finalized mathematical parameters of the model. | `w, b = model.layers[0].get_weights()` |

---

### **Quick Example: The "Learned Math"**

If your training data follows the pattern $y = 2x + 1$:
1. **Before Training:** $w$ and $b$ are random (e.g., $w=0.01, b=0$). Prediction for $x=5$ is $0.05$.
2. **After Training:** $w \approx 2.0$ and $b \approx 1.0$. Prediction for $x=5$ is $11.0$.

# Conclusions: Linear vs. Logistic Regression in Neural Networks

After comparing traditional machine learning libraries (Scikit-Learn) with Neural Network implementations (TensorFlow/Keras), I have drawn the following conclusions:

### 1. Architectural Equivalence
We have demonstrated that a **Single-Layer Neural Network** can replicate traditional regression and classification models:
* A Neural Network with one neuron and **Linear Activation** is mathematically identical to **Simple Linear Regression**.
* A Neural Network with one neuron and **Sigmoid Activation** is mathematically identical to **Logistic Regression**.



### 2. Consistency of Results
The results from `sklearn.linear_model` and `tf.keras.Sequential` are consistent. Both approaches converge on similar weights ($w$) and biases ($b$) when trained on the same dataset, proving that Neural Networks are a generalized framework that encompasses traditional statistical models.

### 3. Flexibility of Neural Networks
Neural Networks offer a unified approach to problem-solving. We can solve both **Regression** (predicting continuous values) and **Classification** (predicting discrete categories) by simply swapping the **Activation Function** and the **Loss Function**:
* **Regression:** `activation='linear'` and `loss='mse'`
* **Classification:** `activation='sigmoid'` and `loss='binary_crossentropy'`



### 4. Scalability
While Scikit-Learn is excellent for simple, small-scale problems, the Neural Network approach (using TensorFlow/Keras) allows us to scale these "simple" models into **Deep Learning** by easily adding hidden layers and non-linearities to solve high-complexity real-world problems.
